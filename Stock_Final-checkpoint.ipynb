```python
import mplfinance as mpf
import pandas as pd
import matplotlib as matplotlib
matplotlib.use("TkAgg")

from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk
from tkcalendar import Calendar,DateEntry


def getCandlestickChart():
    try:
        inputValue=textbox.get("1.0","end-1c")
        rvalue=var.get()
        ticker=inputValue
        file="D:\College\\Northern Trust Hackathon\StockList.json"
        d = pd.read_json(file)
        newdata = d[d['symbol'] == ticker]

        newdata = newdata.set_index('date')
        df = pd.DataFrame(data=newdata)
        chartData = df.sort_values(by='date')
        
        if rvalue==1:
            fig, ax = mpf.plot(chartData,type='candle',
                      volume = True,
                      style = "yahoo",
                      mav=(50),
                      tight_layout=True,
                      figratio=(12,6),
                      figscale=0.50,
                      returnfig=True
                      )
            canvas = FigureCanvasTkAgg(fig,master=graphFrame)   
            canvas.draw()
        elif rvalue==2:
            fig, ax = mpf.plot(chartData,type='line',
                      mav=(50),
                      volume = True,
                      tight_layout=True,
                      figratio=(12,6),
                      figscale=0.50,
                      returnfig=True
                      )
            canvas = FigureCanvasTkAgg(fig,master=graphFrame)   
            canvas.draw()
        elif rvalue==3:
            fig, ax = mpf.plot(chartData,type='candle',
                      mav=(50),
                      tight_layout=True, 
                      figratio=(12,6),
                      figscale=0.50,
                      returnfig=True
                      )
            canvas = FigureCanvasTkAgg(fig,master=graphFrame)   
            canvas.draw()
            
    except IndexError:
        label=tk.Label(window,text="Not Found")
        label.pack(pady=10)



#Setup Tkinter and frames ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
window = tk.Tk()
def toggle_fs(dummy=None):
    state = False if window.attributes('-fullscreen') else True
    window.attributes('-fullscreen', state)
    if not state:
        window.geometry('300x300+100+100')
#window.attributes('-fullscreen', True)
window.title("Plot Test") #Window Title
window.geometry('600x390') #Window Size
#set window color
window.configure(bg='White')

window.attributes('-fullscreen', True)

window.bind('<Escape>', toggle_fs)
#Graph frame
graphFrame = tk.Frame(window, width=0, height= 0)
graphFrame.pack()
#Label~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

label1 = tk.Label(window,text="OHLC Engine DashBoard",fg='#00008B',relief = "solid",font=("arial",30,"bold")).pack(pady=30)

#text~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
label2 = tk.Label(window,text="Search for the Company Symbol",fg='orange',bg="White",font=("arial",20)).pack(pady=10)
textbox=tk.Text(window,width=10,height=1,fg='#00008B',relief = "solid",font=("arial",20,"bold"))
textbox.pack(pady=20)


#Radio Button~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
var = tk.IntVar()
label2 = tk.Label(window,text="Select the Chart Type",fg='orange',bg="White",font=("arial",20)).pack(pady=10)
R1 = tk.Radiobutton(window, text="CandleStick Chart", variable=var, value=1,font=("arial",15))
R1.pack(pady=10)

R2 = tk.Radiobutton(window, text="Line Chart", variable=var, value=2,font=("arial",15))
R2.pack(pady=2)

R3 = tk.Radiobutton(window, text="Hollow Candle", variable=var, value=3,font=("arial",15))
R3.pack(pady=10)

#Add Button ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
btn = tk.Button(window, text="Plot!", command=getCandlestickChart,font=("arial",20))
btn.pack(pady=10)
#text~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
label3 = tk.Label(window,text="ESC to exit",fg='red',bg="White",font=("arial",10)).pack(pady=10)

#Tkinter Main Loop
window.mainloop()
```
